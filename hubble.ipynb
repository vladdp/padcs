```python
import numpy as np

from padcs.satellite import Satellite
```

```python
sim_time = 100

hubble = Satellite(sim_time)

J = [ [36046, -706, 1491],
      [-706, 86868, 449],
      [1491, 339, 93848] ]

q_des = [0, 0, 0, 1] 
q_initial = np.array( [10, 20, 30, 1.0] )
q_initial /= np.linalg.norm(q_initial)

k = 0.05
d = 0.316

mu = 0.9
w_0 = [0.01, 0.01, 0.01]
```

```python
hubble.set_inertia_matrix(J)
hubble.set_gains(k, d)
hubble.set_ang_vel(w_0)
hubble.to_attitude_wie(q_initial, q_des, mu=mu)
```
