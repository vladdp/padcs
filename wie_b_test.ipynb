Using the library to produce the same results from [1].

```python
import numpy as np

from padcs.satellite import Satellite
```

Set the initial values

```python
sat = Satellite(100)

J = [ [1200, 100, -200],
      [100, 2200, 300],
      [-200, 300, 3100]]

q_des = np.array( [0, 0, 0, 1] )
q_initial = np.array( [0.57, 0.57, 0.57, 0.159] )

k = 0.05
d = 0.316

mu = 0.9
w_0 = [0.01, 0.01, 0.01]
```

Assign the values to the satellite

```python
sat.set_inertia_matrix(J)
sat.set_gains(k, d)
sat.set_ang_vel(w_0)
sat.to_attitude(q_initial, q_des, mu=mu)
```

Plot the results

```python
sat.plot_q()
```
