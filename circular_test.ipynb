```python
import numpy as np
import matplotlib.pyplot as plt

from padcs.satellite import Satellite
```

Create Satellite object and set orbital parameters

```python
sim_time = 10000

sat = Satellite(sim_time)

a = 7000
e = 0
i = 0
raan = 0
w = 0
nu = 0

sat.set_orbital_elements(a, e, i, raan, w, nu)
```

Calculate satellite position and velocity during sim_time and plot

```python
sat.set_pos()

fig, ax = plt.subplots(1, 3, figsize=(18, 6))

ax[0].set_title('x, y, z values over time')
ax[0].plot(sat.pos[:, 0], label='x')
ax[0].plot(sat.pos[:, 1], label='y')
ax[0].plot(sat.pos[:, 2], label='z')
ax[0].legend()

ax[1].set_title('2d orbit planes')
ax[1].plot(sat.pos[:, 0], sat.pos[:, 1], label='xy')
ax[1].plot(sat.pos[:, 0], sat.pos[:, 2], label='xz')
ax[1].plot(sat.pos[:, 1], sat.pos[:, 2], label='yz')
ax[1].axis('equal')
ax[1].legend()

ax[2].set_title('velocity components')
ax[2].plot(sat.vel[:, 0], label='v_x')
ax[2].plot(sat.vel[:, 1], label='v_y')
ax[2].plot(sat.vel[:, 2], label='v_z')
ax[2].legend()
```
